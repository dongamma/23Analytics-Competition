```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일을 pandas DataFrame으로 읽어들입니다.
df = pd.read_csv('/content/drive/MyDrive/23년도 2학기 학술제/11-29/TripB28.csv', encoding='cp949')

fig, axs = plt.subplots(8, 6, figsize=(20, 15))  # 4행 7열의 subplot을 생성합니다.
axs = axs.ravel()  # axs를 1차원 배열로 만듭니다.

for i, column in enumerate(df.columns):
    axs[i].plot(df[column])
    axs[i].set_title(column)
    axs[i].grid(True)

plt.tight_layout()  # 그래프 간의 간격을 자동으로 조정합니다.
plt.show()
```
